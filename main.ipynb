```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
```

```python
hex_color_code = [
"#FF0000",
"#00FF00",
"#0000FF",
"#FFFF00",
"#800080"
]
```

```python
x1 = np.linspace(10000, 40000, 30)
x1_coefficient = 5 
x1_intercept = 4000
y1 = x1*x1_coefficient + x1_intercept 
```

```python
plt.figure(figsize=(16, 8))

plt.xlim(3000, 9000)
plt.ylim(0, 55000)

plt.scatter(x1, y1, color=hex_color_code[1], marker='x', label="arm1")
# plt.plot(x2_nc, m2*np.log10(x2_nc) + t2, color=hex_color_code[1])

# plt.scatter(nc_arm1_x,nc_arm1_y, color=hex_color_code[2], marker='o', label="arm2")
# plt.plot(x1_nc, m1*np.log10(x1_nc) + t1, color=hex_color_code[2],)

plt.xlabel('Weekly Spend', fontsize=18)
plt.ylabel('Weekly Transaction Value', fontsize=18)
plt.legend(loc="upper left", fontsize=18)

xlabels = ['€'+'{:,.0f}'.format(x) + 'K' for x in plt.gca().get_xticks()/1000]
plt.gca().set_xticklabels(xlabels)

ylabels = ['€'+'{:,.0f}'.format(y) + 'K' for y in plt.gca().get_yticks()/1000]
plt.gca().set_yticklabels(ylabels)

plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

plt.title('Transaction Value by Spend', fontsize=24)
plt.savefig('viz/NCTrxValuebySpend.png', dpi=300, bbox_inches='tight')
plt.show()
```
